```python
import openai
import time
key = "sk-Cj2lZwA_uk-IP6RzpyFb8nLWvn06TBliPojsHnP2_vT3BlbkFJJ_rK_Pp-d1PFIdInoRN0GInDvEVlLu-3lb3td0N-cA"

# Function to get latitude and longitude using ChatGPT API
def get_coordinates(site_name, region, state, country="Mexico"):
    client = openai.OpenAI(api_key=key)  # Replace with your OpenAI API key
    prompt = f"Find the latitude and longitude of {site_name} in {region}, {state}, {country}. Print as latitude: x, longitude: y"
    
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": "You are a helpful assistant."},
                  {"role": "user", "content": prompt}]
    )
    
    answer = response.choices[0].message.content
    return answer

```

```python
data['response'] = np.nan
# Fetch coordinates for each site
for i,site in data.iterrows():
    coordinates = get_coordinates(site["Site"], site["Region"], site["State/Privince"])
    data.loc[i, "response"] = coordinates
    print(f"{site['Site']} ({site['State/Privince']}): {coordinates}")
    try:
        coordinates_list = [float(c[:-1]) for c in coordinates.split(" ") if c[:-1].replace('.', '', 1).isdigit()]
        data.loc[i, "Latitude"] = coordinates_list[0]
        data.loc[i, "Longitude"] = coordinates_list[1]
    except:
        print(f"Error fetching coordinates for {site['Site']} ({site['State/Privince']})")
    time.sleep(1)  # Sleep to avoid overwhelming the API
```
